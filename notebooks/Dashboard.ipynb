```python
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
from dash import Dash , html , dcc,callback
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../data/processed/datset_cleaned.csv')
df
```

```python
fig3=px.pie(names=df.groupby('event_location_region')['place_of_residence'].nunique().index,values=df.groupby('event_location_region')['place_of_residence'].nunique().values,title='event_location_region Vs place_of_residence')
fig3.show()
```

```python
fig=px.line(df.groupby(['event_location_region','year']).name.count().reset_index(),x='year',y='name',color='event_location_region')
fig.update_layout(legend={
    "title":'event_location_region',
    'yanchor':'top',
    'xanchor':'right'
})
fig.show()
```

```python
fig1=px.funnel(df.event_location_district.value_counts().reset_index(),y='event_location_district' ,x='count')
fig1.update_layout(legend={
    "title":'event_location_district',
})
fig1.show()
```

```python
fig4=px.pie(names=df.event_location_region.value_counts().index,values=df.event_location_region.value_counts().values,title='West Bank Vs. Gaza Strip Vs. Israel')
fig4.show()
```

```python
fig2=px.bar(df['event_location_region'].value_counts(),y='count',title='Count of Event Location Region')
fig2.show()
```

```python
fig5=px.bar(df['gender'].value_counts(),y='count',title='NO.Gender')
fig5.show()
```

```python
fig6=px.bar(df['type_of_injury'].value_counts(),y='count',title='Count Of Type Of Injury')
fig6.show()
```

```python
fig7=px.pie(names=df.killed_by.value_counts().index,values=df.killed_by.value_counts().values,title='Israeli security forces Vs. Palestinian civilians Vs. Israeli civilians')
fig7.show()
```

```python
fig8=px.pie(names=df.type_of_injury.value_counts().index,values=df.type_of_injury.value_counts().values)
fig8.show()
```

```python
fig9=px.pie(names=df.citizenship.value_counts().index,values=df.citizenship.value_counts().values)
fig9.show()
```

```python
fig11=px.scatter(df,x='date_of_death',y='name',color='gender',color_discrete_map={'F':'blue','M':'red'})
fig11.update_layout(legend={
    "title":'Gender'
})
fig11.show()
```

```python
fig10=px.scatter(df,x='citizenship',y='event_location',color='gender',color_discrete_map={'F':'blue','M':'red'})
fig10.update_layout(legend={
    "title":'Gender',
    'yanchor':'top',
    'xanchor':'right'
})
fig10.show()
```
